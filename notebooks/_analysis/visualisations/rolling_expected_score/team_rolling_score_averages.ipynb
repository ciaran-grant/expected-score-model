Plot all teams

```python
team_list = list(team_colours.keys())
team_list
```

```python
fig = plt.figure(figsize=(18, 24), dpi=300)

nrows=6
ncols=3
gspec = gridspec.GridSpec(
    ncols=ncols, nrows=nrows, figure=fig,
    hspace=0.3
)
plot_counter = 0
for row in range(nrows):
    for col in range(ncols):
        team = team_list[plot_counter]
        colour_for = team_colours[team]['positive']
        colour_against = team_colours[team]['negative']
        ax = plt.subplot(gspec[row, col])
        get_actual_rolling_plot(ax, team, data=df_shots, window=10, color_for=colour_for, color_against=colour_against)
        plot_counter+=1

fig_text(
    x=0.13, y=0.92,
    s = "Score For & Against - 10-game rolling average.",
    size = 22,
    font = "Karla"
)
fig_text(
    x=0.13, y=0.905,
    s = "For & Against values are @ end of 2022.",
    size = 14,
    font = "Karla"
)

plt.savefig(
	"figures/20230718_afl_rolling_score.png",
	dpi = 600,
	facecolor = "#121212",
	bbox_inches="tight",
    edgecolor="none",
	transparent = False
)

plt.savefig(
	"figures/20230718_afl_rolling_score_tr.png",
	dpi = 600,
	facecolor = "none",
	bbox_inches="tight",
    edgecolor="none",
	transparent = True
)
```
