Expected Score - Set Shot

```python
pitch = VerticalPitch(pitch_width=PITCH_WIDTH, pitch_length=PITCH_LENGTH, 
                    line_colour="white", pitch_colour="#121212", 
                    line_width=0.5,
                    half = True, pad_bottom=-5)
expected_goal_set_probs = get_expected_pitch_probability(pitch, outcome="goal", shot_type="set", calibrated=True, initial_state="ball_up")
expected_behind_set_probs = get_expected_pitch_probability(pitch, outcome="behind", shot_type="set", calibrated=True, initial_state="ball_up")
expected_miss_set_probs = get_expected_pitch_probability(pitch, outcome="miss", shot_type="set", calibrated=True, initial_state="ball_up")

expected_goal_set_probs_norm = expected_goal_set_probs / (expected_goal_set_probs + expected_behind_set_probs + expected_miss_set_probs)
expected_behind_set_probs_norm = expected_behind_set_probs / (expected_goal_set_probs + expected_behind_set_probs + expected_miss_set_probs)
expected_miss_set_probs_norm = expected_miss_set_probs / (expected_goal_set_probs + expected_behind_set_probs + expected_miss_set_probs)
expected_score = 6*expected_goal_set_probs_norm + expected_behind_set_probs_norm

fig, ax = pitch.draw()
im = ax.imshow(expected_score, vmin=0, vmax=6, cmap="magma", interpolation="gaussian", extent=(-pitch.dim.pitch_width/2, pitch.dim.pitch_width/2, 0, pitch.dim.pitch_length/2), origin="lower", zorder=0)
clip_image_to_pitch_boundary(pitch, ax=ax, im=im)

fig_text(0.1, 0.80, s="Expected Score - Set Shots", font="Karla", fontsize=16)

plt.savefig(
	"figures/20230726_expected_score_set_shots.png",
	dpi = 600,
	facecolor = "#121212",
	bbox_inches="tight",
    edgecolor="none",
	transparent = False
)
```
