Plot Chain on Pitch

```python
import pandas as pd
import numpy as np
from mplfooty.pitch import Pitch
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
pd.options.display.max_rows = 100
pd.options.display.max_columns = 999
```

Load Data

```python
df_chain = pd.read_csv("/Users/ciaran/Documents/Projects/AFL/git-repositories/expected-score-model/data/match_chains.csv")
df_chain.head()
```

Get Chain

```python
match_id = "202101_BrisbaneLions_Sydney"
chain_number = 1
```

```python
chain = df_chain[(df_chain['Match_ID'] == match_id) & (df_chain['Chain_Number'] == chain_number)]
```

Plot Chain

```python
def get_venue_dimensions(chain_data, match_id):
    
    # Get match chain information
    match = chain_data[chain_data['Match_ID'] == match_id]

    return list(set(match['Venue_Length']))[0], list(set(match['Venue_Width']))[0]
```

```python
pitch_length, pitch_width = get_venue_dimensions(df_chain, match_id)
```

```python
chain
```

```python
pitch = Pitch(pitch_width=pitch_width, pitch_length=pitch_length, line_colour="grey")
fig, ax = pitch.draw()
pitch.plot(chain['x'], chain['y'], ax=ax)
pitch.scatter(chain['x'], chain['y'], ax=ax, color="r")
```
